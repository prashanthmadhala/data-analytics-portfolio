```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import pandas as pd

# Select relevant features
features = df[['freight_value', 'payment_value', 'payment_sequential', 'payment_installments']]
```

```python
# Normalize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Find the optimal number of clusters using the Elbow Method
wcss = []  # Within-cluster sum of squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method to Determine Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within-Cluster Sum of Squares)')
plt.show()
```

```python
# Find the optimal number of clusters using Silhouette Score
silhouette_scores = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_features)
    silhouette_avg = silhouette_score(scaled_features, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the Silhouette Scores
plt.figure(figsize=(8, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score to Determine Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

```python
# Choose the optimal number of clusters (based on the plots above)
optimal_clusters = 3  # Adjust this based on the elbow method and silhouette score
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)
```

```python
# Analyze the clusters
print('Cluster centers (scaled values):')
print(kmeans.cluster_centers_)
```

```python
print('\nNumber of customers in each cluster:')
print(df['cluster'].value_counts())
```

```python
from scipy.stats import f_oneway

# Variables to test
variables = ['freight_value', 'payment_value', 'payment_sequential', 'payment_installments']

# ANOVA for each variable
for var in variables:
    anova_result = f_oneway(*(df[df['cluster'] == cluster][var] for cluster in df['cluster'].unique()))
    print(f"ANOVA for {var}:")
    print(f"F-statistic: {anova_result.statistic:.4f}, P-value: {anova_result.pvalue:.4f}")
    if anova_result.pvalue < 0.05:
        print("=> Significant differences exist between clusters for this variable.\n")
    else:
        print("=> No significant differences between clusters for this variable.\n")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Apply Tukey's HSD for each variable
for var in variables:
    tukey = pairwise_tukeyhsd(endog=df[var], groups=df['cluster'], alpha=0.05)
    print(f"Tukey's HSD Test for {var}:")
    print(tukey)
    print("\n")
```

```python
# Freight costs by customer state
freight_by_customer_state = df.groupby('customer_state')['freight_value'].mean().sort_values(ascending=False)
print("Average freight cost by customer state:")
print(freight_by_customer_state)

# Freight costs by seller state
freight_by_seller_state = df.groupby('seller_state')['freight_value'].mean().sort_values(ascending=False)
print("Average freight cost by seller state:")
print(freight_by_seller_state)

# Freight costs by customer-seller city pair
freight_by_city_pair = df.groupby(['customer_city', 'seller_city'])['freight_value'].mean().sort_values(ascending=False)
print("Top customer-seller city pairs with the highest freight cost:")
print(freight_by_city_pair.head(10))
```

```python
import matplotlib.pyplot as plt

# Freight cost by customer state
freight_by_customer_state.plot(kind='bar', figsize=(12, 6))
plt.title('Average Freight Cost by Customer State')
plt.xlabel('Customer State')
plt.ylabel('Average Freight Cost')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Freight cost by seller state
freight_by_seller_state.plot(kind='bar', figsize=(12, 6), color='orange')
plt.title('Average Freight Cost by Seller State')
plt.xlabel('Seller State')
plt.ylabel('Average Freight Cost')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Average freight cost by seller
freight_by_seller = df.groupby('seller_id')['freight_value'].mean().sort_values(ascending=False)
print("Top sellers with the highest average freight cost:")
print(freight_by_seller.head(10))

# Average freight cost by product
freight_by_product = df.groupby('product_id')['freight_value'].mean().sort_values(ascending=False)
print("Top products with the highest average freight cost:")
print(freight_by_product.head(10))
```

```python
# Statistical summary for freight cost by seller
freight_seller_stats = freight_by_seller.describe()
print("Statistical summary of freight cost by seller:")
print(freight_seller_stats)

# Statistical summary for freight cost by product
freight_product_stats = freight_by_product.describe()
print("\nStatistical summary of freight cost by product:")
print(freight_product_stats)
```

```python
from scipy.stats import zscore

# Calculate Z-scores for freight cost by seller
freight_by_seller_zscores = zscore(freight_by_seller)
outliers_seller = freight_by_seller[freight_by_seller_zscores > 3]  # Threshold Z > 3
print("\nSellers with disproportionately high freight costs (Z > 3):")
print(outliers_seller)

# Calculate Z-scores for freight cost by product
freight_by_product_zscores = zscore(freight_by_product)
outliers_product = freight_by_product[freight_by_product_zscores > 3]  # Threshold Z > 3
print("\nProducts with disproportionately high freight costs (Z > 3):")
print(outliers_product)
```

```python
# IQR for freight cost by seller
Q1_seller = freight_by_seller.quantile(0.25)
Q3_seller = freight_by_seller.quantile(0.75)
IQR_seller = Q3_seller - Q1_seller

outliers_seller_iqr = freight_by_seller[freight_by_seller > (Q3_seller + 1.5 * IQR_seller)]
print("\nSellers with disproportionately high freight costs (IQR method):")
print(outliers_seller_iqr)

# IQR for freight cost by product
Q1_product = freight_by_product.quantile(0.25)
Q3_product = freight_by_product.quantile(0.75)
IQR_product = Q3_product - Q1_product

outliers_product_iqr = freight_by_product[freight_by_product > (Q3_product + 1.5 * IQR_product)]
print("\nProducts with disproportionately high freight costs (IQR method):")
print(outliers_product_iqr)
```
